

Upon examining the dataset, several columns have incorrect data types that need to be updated to ensure proper analysis and visualization. Here are the necessary adjustments:  

1. **Reviews**:  
   - **Current Data Type**: Object  
   - **Required Data Type**: Integer or Float  
   - **Reason**: The number of reviews should be a numerical value for aggregation and statistical analysis.  

2. **Size**:  
   - **Current Data Type**: Object  
   - **Required Data Type**: Float or Integer  
   - **Reason**: The app size, often in KB or MB, is numerical data and needs to be converted to facilitate comparisons and calculations.  

3. **Installs**:  
   - **Current Data Type**: Object  
   - **Required Data Type**: Integer  
   - **Reason**: Install count is a numeric value and should be converted to enable meaningful sorting, filtering, and analysis.  

4. **Price**:  
   - **Current Data Type**: Object  
   - **Required Data Type**: Float  
   - **Reason**: Prices, even in string format (e.g., "$4.99"), need to be converted into numerical values for computations. The dollar symbol can be stripped during preprocessing.  

5. **Last Updated**:  
   - **Current Data Type**: Object  
   - **Required Data Type**: DateTime  
   - **Reason**: Dates are crucial for time-based analysis and trends. Converting this to a DateTime format ensures proper sorting and time-related calculations.  



These changes will enhance the dataset's usability and ensure more accurate and meaningful analyses.

````python
# Convert the column back to object
df['Reviews'] = df['Reviews'].astype(str)

print('Before converting data type of Reviews column:', df['Reviews'].dtype)

# Function to convert reviews
def convert_reviews(reviews):
    reviews = str(reviews).strip()

    if 'M' in reviews:
        return float(reviews.replace('M', '')) * 1000000
    elif 'K' in reviews:
        return float(reviews.replace('K', '')) * 1000
    else:
        return float(reviews)

# Apply the function
df['Reviews'] = df['Reviews'].apply(convert_reviews)

# Convert to integer
df['Reviews'] = df['Reviews'].astype(int)

print('After converting data type of Reviews column:', df['Reviews'].dtype)
````

````python
# Convert the column back to object (string) for demonstration
df['Size (MB)'] = df['Size (MB)'].astype(str)

print('Before converting datatype of Size column:', df['Size (MB)'].dtype)

# Function to convert Size to MB
def convert_size(size):
    size = str(size).strip()

    if 'm' in size.lower():
        return float(size.lower().replace('m', '').replace(',', ''))

    elif 'k' in size.lower():
        return float(size.lower().replace('k', '').replace(',', '')) / 1024

    elif '+' in size:
        return float(size.replace(',', '').replace('+', ''))

    elif size.lower() == 'varies with device':
        return np.nan

    else:
        return np.nan

# Apply conversion
df['Size (MB)'] = df['Size (MB)'].apply(convert_size)

print('After converting datatype of Size column:', df['Size (MB)'].dtype)
````

````python
# Converting Installs column dataype to numeric

def convert_installs(installs):
    # If the value is already a number, return it directly as an integer
    if isinstance(installs, (int, float)):
        return int(installs)
    # Ensure it's a string for the following checks
    installs = str(installs)
    # Handle 'Free' case and replace it with 0
    if installs.lower() == 'free':
        return 0
    # Remove the comma and the '+' symbol, convert to integer
    elif '+' in installs:
        return int(installs.replace(',', '').replace('+', ''))
    # Handling '0+' or '0' case as 0 installs
    elif installs == '0+' or installs == '0':
        return 0
    # Otherwise, removing the comma and return as integer
    else:
        return int(installs.replace(',', ''))

print('Before coverting data type of installs column:',df['Installs'].dtypes)
df['Installs'] = df['Installs'].apply(convert_installs) # Appling the function to the 'Installs' column
print('After coverting data type of installs column:',df['Installs'].dtypes)
````

````python
# Converting Price column dataype to numeric

df['Price ($)'] = df['Price ($)'].astype(str)

print('Before converting data type of Price column:', df['Price ($)'].dtype)

# Function to process price
def process_price(price):
    if price == '0':
        return 0.0
    elif price.startswith('$'):
        return float(price.replace('$', ''))
    else:
        return np.nan

# Convert to numeric
df['Price ($)'] = df['Price ($)'].apply(process_price)

print('After converting data type of Price column:', df['Price ($)'].dtype)
````

````python
# Convert 'Last Updated' column to datetime, coercing errors
# Convert datetime column to string (object)
df['Last Updated'] = df['Last Updated'].astype(str)

print('Before converting data type of Last Updated column:', df['Last Updated'].dtype)

# Convert back to datetime
df['Last Updated'] = pd.to_datetime(df['Last Updated'], errors='coerce')

print('After converting data type of Last Updated column:', df['Last Updated'].dtype)
````

````python
# Fill missing values in the 'Type' column with the mode (most frequent value)
# 'Type' typically indicates whether the app is 'Free' or 'Paid', so filling with the mode is appropriate.
df['Type'] = df['Type'].fillna(df['Type'].mode()[0])

# Fill missing values in the Price ($) column with the median value
# The median is robust to outliers, making it suitable for numerical columns like price, especially if skewed.
df['Price ($)'] = df['Price ($)'].fillna(df['Price ($)'].median())

# Fill missing values in the Content Rating column with the mode
# 'Content Rating' indicates the target audience for the app (e.g., 'Everyone', 'Teen').
# Using the mode ensures consistency with the most common value.
df['Content Rating'] = df['Content Rating'].fillna(df['Content Rating'].mode()[0])

# Fill missing values in the Last Updated column with the mode
# Last Updated represents the last update date of the app.
# Filling with the mode is reasonable if missing values are minimal, assuming most apps share a common recent update date.
df['Last Updated'] = df['Last Updated'].fillna(df['Last Updated'].mode()[0])

# Fill missing values in the 'Current Ver' column with the mode
# 'Current Ver' indicates the app version. Using the mode assumes the most common version is representative of missing entries.
df['Current Ver'] = df['Current Ver'].fillna(df['Current Ver'].mode()[0])

# Fill missing values in the 'Android Ver' column with the mode
# 'Android Ver' indicates the minimum Android version required. The mode represents the most common compatibility level.
df['Android Ver'] = df['Android Ver'].fillna(df['Android Ver'].mode()[0])
````

````python
# Creating a new column - Install_Category

# Define bins and labels based on Installs
bins = [-1, 0, 1000, 10000, 100000, 1000000, 10000000, float('inf')]
labels = ['Zero Installs', 'Very Low', 'Low', 'Moderate', 'High', 'Very High', 'Extremely High']

# Create the new column
df.loc[:, 'Install_Category'] = pd.cut(df['Installs'], bins=bins, labels=labels, right=False)

# Display the first few rows
df[['Installs', 'Install_Category']].head()
````

- To simplify analysis by grouping installation numbers into categories.

Dropping the rows where both `Translated_Review` and `Sentiment` are missing for the same `App` is essential for the following reasons:

2. **Data Integrity**: Ensuring that each app has at least some form of review and sentiment information helps maintain the integrity of the dataset. Rows with missing values in these key fields do not contribute to any analysis, which could distort any findings or trends.

3. **Accurate Analysis**: For tasks like sentiment analysis, review-based recommendations, or app performance assessment, missing review or sentiment data for certain apps would lead to skewed results. Dropping these rows ensures that the analysis is based on complete and accurate information.

4. **Better Data Representation**: By removing rows where the app's review and sentiment data are entirely missing, we focus on the apps that have valuable user input. This helps in building more accurate models, visualizations, or reports that represent actual user sentiments.

Dropping these rows ensures that the dataset remains clean, accurate, and relevant for meaningful analysis and decision-making.

2. **'Sentiment_Polarity' and 'Sentiment_Subjectivity' Columns:**

This approach ensures that missing values are imputed contextually within each `App` group, maintaining the dataset's integrity and consistency for downstream analysis. It also minimizes data loss and prevents bias introduced by indiscriminate imputation methods.
